```python
from bs4 import BeautifulSoup
import requests

url='https://www.wetter.de/deutschland/wetter-frankfurt-18221009.html'
website = requests.get(url)
data = BeautifulSoup(website.content, 'html.parser')

#daily temperatures
temperature_max = int(data.find('div', class_='weather-daybox-base__minMax__max').text.strip())
temperature_min = int(data.find('div', class_='weather-daybox-base__minMax__min').text.strip())
print('Maximum temperature for today: ' + str(temperature_max) + '째C.')
print('Minimum temperature: ' + str(temperature_min) + '째C.')

#further hourly weather data, added to list
weatherbox_results = []
weatherbox = data.find('dl', 'weather-daybox__main__hourInfos')
for element in weatherbox:
    element = weatherbox.find_all('dd')
    weatherbox_results.append(element)

rain_probability = int(weatherbox_results[0][0].text.strip()[0:-26])
print('Rain probability: ' + str(rain_probability) + '%.')

#wind = int(information_results[0][1].text.strip()[0:-21])
#print('Strength of wind: ' + str(wind) + 'km/h.')

#Pulling sun time, checking if given number is in minutes or hours.
sun = weatherbox_results[0][3].text.strip()
if 'Min' in sun:
    hours_mins = 'minute/s'
else:
    hours_mins = 'hour/s'    
sun_time = int(sun[0:-18])
print('Sun: ' + str(sun_time) + ' ' + str(hours_mins) + '.')


#recommendation: Clothes to wear today
print('=========================')
print('=========================')

print('Today you should wear:')

if rain_probability > 50 and temperature_min > 0:
    print('umbrella')
    print('solid shoes')
    
if rain_probability > 50 and temperature_min < 0:
    print('solid shoes')
    
if rain_probability > 50 and temperature_max > 5:
    print('rain jacket')

if temperature_max <= 5:
    print('winterjacket')
    print('scarf')

if temperature_max > 5 and temperature_min < 5:
    print('if you are out at night/in evening: winterjacket')
    print('if you are out at night/in evening: scarf')
    
if temperature_max <= 0:
    print('long underwear')
    print('gloves')

if temperature_max > 5 and temperature_max < 16:
    print('transitional jacket')

if temperature_max <= 18:
    print('sweater')
    
if temperature_max > 18:
    print('t-shirt')
    
if temperature_max > 25:
    print('shorts')
    
if temperature_max > 0 and temperature_min <= 0:
    print('if you are out at night/in evening: long underwear')
    print('if you are out at night/in evening: gloves')

if sun_time > 5 and hours_mins == 'hour/s':
    print('sunglasses')
    
#if wind >= 60 and wind < 88:
#    print('Attention: Stronger winds! Be careful')

#if wind >= 88:
#    print('WARNING: Dangerous winds, stay home!')

if rain_probability > 50 and temperature_min <0:
    print('Attention: Snowfall possible.')

```
